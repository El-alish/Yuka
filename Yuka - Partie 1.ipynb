# Mission 4

Quel est le nombre de colonne appartenant à chaque quartile des taux de remplissage ? Autrement dit, combien de colonne sont remplies à 0%, entre 0 et 25%, entre 25 et 50%...

```python
import seaborn as sns

fill_rates = pd.DataFrame({'Colonnes': df.columns, 'Taux de remplissage': df.count() / len(df)}) #on crée un dataframeexpres avec les colonnes et le taux de remplissage

quartiles = ['0-25%', '25-50%', '50-75%', '75-100%'] # Créer une liste des quartiles

# Calculer le nombre de colonnes dans chaque quartile
counts = [sum(fill_rates['Taux de remplissage'].between(i, j)) for i, j in [(0, 0.25), (0.25, 0.5), (0.5, 0.75), (0.75, 1)]]

# Afficher les résultats
for quartile, count in zip(quartiles, counts):
    print(f"Nombre de colonnes dans le quartile {quartile}: {count}")
fill_rates['Taux de remplissage'].plot(kind='hist', bins=30, edgecolor='black', color='skyblue')
plt.title('Distribution des taux de remplissage des colonnes')
plt.xlabel('Taux de remplissage')
plt.ylabel('Nombre de colonnes')
plt.show()
print(fill_rates.sort_values(by='Taux de remplissage'))
```

# Mission 6

Un dataset trop pauvre en data n'est pas vraiment intéressant. Après avoir analyser les colonnes, nous nous sommes rendu compte que nous pouvons supprimer les colonnes avec moins de 60% de valeurs non nulles, sauf les colonnes pnns_groups_1 et pnns_groups_2, nutrition-score-fr_100g et nutrition-score-uk_100g. En effet ces 2 colonnes vont nous être utiles par la suite.

Si ce n'est pas déjà fait, supprime également les colonnes suivantes car elles ne sont pas vraiment utiles pour répondre à notre problématique :
* creator
* created_t
* created_datetime
* last_modified_t
* brands_tags
* countries
* countries_tags
* additives_tags
* states
* states_tags

```python
dffrance.columns
```

```python


# Calculer le seuil de 60% de valeurs non nulles
valeursnonnulles = len(dffrance) * 0.6

# Colonnes à garder indépendamment du nombre de valeurs non nulles
colonnesagarder= ['pnns_groups_1', 'pnns_groups_2', 'nutrition-score-fr_100g', 'nutrition-score-uk_100g']

# Liste des colonnes à supprimer si elles ont moins de 60% de valeurs non nulles
columns_to_drop = [col for col in dffrance.columns if (dffrance[col].notna().sum() < valeursnonnulles) and (col not in colonnesagarder)]

# Supprimer les colonnes identifiées
df_cleaned = dffrance.drop(columns=columns_to_drop)

print("Colonnes supprimées:", columns_to_drop)
df_cleaned
```
